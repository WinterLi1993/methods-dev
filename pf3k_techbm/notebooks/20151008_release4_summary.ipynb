# Load data

```python
ref_dict=SeqIO.to_dict(SeqIO.parse(open(REF_GENOME), "fasta"))
ref_dict.keys()
```

```python
reload = False
if not os.path.exists(ALL_VARIANTS_FN) or reload:
    npy_arrays = []
    for i, chrom in enumerate(ref_dict.keys()):
        print(i, chrom)
        npy_fn = FULL_NPY_FORMAT % chrom
        npy_arrays.append(np.load(npy_fn))
    all_variants = np.concatenate(npy_arrays)
    np.save(ALL_VARIANTS_FN, all_variants)
else:
    all_variants = np.load(ALL_VARIANTS_FN)

    
```

```python
flt_pass = ((all_variants['RegionType'] == b'Core') & (all_variants['VQSLOD'] > 0))
pass_variants = all_variants[flt_pass]
np.unique(flt_pass, return_counts=True)
```

```python
non_coding_effects = np.array([b'INTERGENIC', b'INTRON'])
snp_types = np.array([b'SNP', b'MULTIALLELIC_SNP'])
```

```python
flt_types = collections.OrderedDict()
flt_types['Coding SNP'] = (
    (np.in1d(pass_variants['VariantType'], snp_types)) &
    np.logical_not(np.in1d(pass_variants['SNPEFF_EFFECT'], non_coding_effects))
)
flt_types['Non-coding SNP'] = (
    (np.in1d(pass_variants['VariantType'], snp_types)) &
    (np.in1d(pass_variants['SNPEFF_EFFECT'], non_coding_effects))
)
flt_types['Coding indel'] = (
    np.logical_not(np.in1d(pass_variants['VariantType'], snp_types)) &
    np.logical_not(np.in1d(pass_variants['SNPEFF_EFFECT'], non_coding_effects))
)
flt_types['Non-coding indel'] = (
    np.logical_not(np.in1d(pass_variants['VariantType'], snp_types)) &
    (np.in1d(pass_variants['SNPEFF_EFFECT'], non_coding_effects))
)
```

```python
tbl_summary = [['Variant type', '# segregating sites', '# non-ref alleles', 'mean NRAF', 'mean NRAF 2', 'mean MAF', 'mean variants per sample']]
nraf_ac = np.sum(pass_variants['AC'], 1)
nraf_af = np.sum(pass_variants['AF'], 1)
maf = 0.5 - np.absolute(nraf_af-0.5)
for flt_type in flt_types:
    tbl_summary.append([
        flt_type,
        np.count_nonzero(flt_types[flt_type]),
        int(np.sum(nraf_ac[flt_types[flt_type]]/2)),
        round(np.mean(nraf_ac[flt_types[flt_type]]) / 5024, 3),
        round(np.mean(nraf_af[flt_types[flt_type]]), 3),
        round(np.mean(maf[flt_types[flt_type]]), 3),
        int(np.sum(nraf_ac[flt_types[flt_type]]) / 5024)
    ])
#     print(
#         flt_type,
#         np.count_nonzero(flt_types[flt_type]),
#         np.sum(nraf_ac[flt_types[flt_type]]),
#         np.mean(nraf_ac[flt_types[flt_type]]) / 5024,
#         np.sum(nraf_ac[flt_types[flt_type]]) / 5024
#     )
```
