```python
crosses_fns = ['../../meta/crosses_samples_%s.txt' % x for x in ['3D7xHB3', 'HB3xDd2', '7G8xGB4']]
crosses_samples = (etl
    .fromtsv(crosses_fns[0])
    .cat(etl.fromtsv(crosses_fns[1]))
    .cat(etl.fromtsv(crosses_fns[2]))
    .convert('sample', 'replace', 'PG0083-C', '7G8')
    .convert('sample', 'replace', 'PG0084-C', 'GB4')
    .sort('sample')
)
release_5_metadata_fn = '../../meta/pf3k_release_5_metadata_20160208.xls'
release_5_metadata = etl.fromxls(release_5_metadata_fn).convertnumbers()

print(len(crosses_samples))
crosses_metadata = (release_5_metadata
    .join(crosses_samples, key='sample')
    .cut(['sample', 'acc', 'study', 'study_title', 'clone'])
    .addfield('parent_or_progeny', lambda rec: 'parent' if rec['clone'] in ['HB3', 'DD2', '3D7', '7G8', 'GB4'] else 'progeny')
    .sort(['study', 'parent_or_progeny', 'clone'])
)
print(len(crosses_metadata))

release5_crosses_metadata_xlsx_fn = '../../meta/pf3k_release_5_crosses_metadata.xlsx'
release5_crosses_metadata_txt_fn = '../../meta/pf3k_release_5_crosses_metadata.txt'
```

```python
crosses_metadata.displayall()
```

```python
crosses_metadata.toxlsx(release5_crosses_metadata_xlsx_fn)
crosses_metadata.totsv(release5_crosses_metadata_txt_fn)
```
