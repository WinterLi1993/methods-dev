# Analysis

```python
array_release4 = np.load(release4_vcfnp_dir['7G8'] + '/variants.npy')
array_release4 = array_release4[['CHROM', 'POS', 'REF', 'ALT', 'is_snp', 'SNPEFF_EFFECT', 'VQSLOD']]
chrom_pos = np.core.defchararray.add(
    array_release4['CHROM'],
    array_release4['POS'].astype('S10')
)
array_release4 = np.lib.recfunctions.append_fields(
    array_release4,
    'chrom_pos',
    chrom_pos
)
    
array_validation5 = np.load(validation5_vcfnp_dir['7G8'] + '/variants.npy')
array_validation5 = array_validation5[['CHROM', 'POS', 'REF', 'ALT', 'is_snp', 'SNPEFF_EFFECT', 'VQSLOD']]
array_validation5.dtype.names = ["val5_%s" % x for x in array_validation5.dtype.names]
chrom_pos = np.core.defchararray.add(
    array_validation5['val5_CHROM'],
    array_validation5['val5_POS'].astype('S10')
)
array_validation5 = np.lib.recfunctions.append_fields(
    array_validation5,
    'chrom_pos',
    chrom_pos
)

flt_array_release4 = np.array([x[0] != '*' for x in array_release4['ALT']])
flt_array_validation5 = np.array([x[0] != '*' for x in array_validation5['val5_ALT']])

# array_release4_filtered = array_release4[array_release4['


array_comparison = np.lib.recfunctions.join_by(
    'chrom_pos', 
    array_release4[flt_array_release4],
    array_validation5[flt_array_validation5]
)
```

```python
(etl.fromarray(array_comparison.data)
.addfield('PASS_rel4', lambda rec: rec['VQSLOD'] > 0)
.addfield('PASS_val5', lambda rec: rec['val5_VQSLOD'] > 0)
.valuecounts('PASS_rel4', 'PASS_val5')
)
```

```python
tbl_release4 = (
    etl.fromarray(np.load(release4_vcfnp_dir['7G8'] + '/variants.npy'))
    .cut(['CHROM', 'POS', 'REF', 'ALT', 'num_alleles', 'is_snp', 'SNPEFF_EFFECT', 'VQSLOD'])
    .select(lambda rec: rec['ALT'][0] != '*')
    .addfield('PASS_release4', lambda rec: rec['VQSLOD'] > 0)
    .addfield('is_in_release4', True)
)
tbl_validation5 = (
    etl.fromarray(np.load(validation5_vcfnp_dir['7G8'] + '/variants.npy'))
    .cut(['CHROM', 'POS', 'REF', 'ALT', 'num_alleles', 'is_snp', 'SNPEFF_EFFECT', 'VQSLOD'])
    .select(lambda rec: rec['ALT'][0] != '*')
    .addfield('PASS_validation5', lambda rec: rec['VQSLOD'] > 0)
    .addfield('is_in_validation5', True)
)
tbl_7g8 = tbl_release4.join(tbl_validation5, key=(['CHROM', 'POS']))
tbl_7g8
```

```python
tbl_release4 = (
    etl.fromvcf(release4_vcfnp_dir['7G8'] + '/../WG.7G8.vcf.gz')
    .vcfmeltsamples()
    .vcfunpackcall()
    .unpackdict('INFO', samplesize=1000000)
    .cut(['CHROM', 'POS', 'REF', 'ALT', 'SNPEFF_EFFECT', 'VQSLOD'])
    .select(lambda rec: rec['ALT'][0] != '*')
    .addfield('PASS_release4', lambda rec: rec['VQSLOD'] > 0)
    .addfield('is_in_release4', True)
)
tbl_validation5 = (
    etl.fromvcf(release4_vcfnp_dir['7G8'] + '/../validation_WG.7G8.vcf.gz')
    .vcfmeltsamples()
    .vcfunpackcall()
    .unpackdict('INFO', samplesize=1000000)
    .cut(['CHROM', 'POS', 'REF', 'ALT', 'SNPEFF_EFFECT', 'VQSLOD'])
    .select(lambda rec: rec['ALT'][0] != '*')
    .addfield('PASS_validation5', lambda rec: rec['VQSLOD'] > 0)
    .addfield('is_in_validation5', True)
)
tbl_7g8 = tbl_release4.join(tbl_validation5, key=(['CHROM', 'POS']))
tbl_7g8
```
