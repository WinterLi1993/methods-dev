```python
for rec in etl.cat(
    tbl_pf_40_freeze_manifest.selecteq('study', 'PF13'),
    tbl_pf_40_freeze_manifest.selectne('study', 'PF13')
).data():
    bam_fn = rec[0]
    ox_code = rec[2].replace('-', '_')
#     print("\n", ox_code, end=":")
    print('.', end="")
    for exon in analysis_exons:
#         print('.', end="")
        exon_length = analysis_exons[exon][2] - analysis_exons[exon][1]
        output_fn = "%s/results/%s/%s_%s.coverage_%dbp.txt" % (output_dir, exon, ox_code, exon, exon_length)
        script_fn = "%s/scripts/%s/%s_%s.coverage_%dbp.txt" % (output_dir, exon, ox_code, exon, exon_length)
        log_filestem = "%s/logs/%s/%s_%s.coverage_%dbp" % (output_dir, exon, ox_code, exon, exon_length)
        if not os.path.exists(os.path.dirname(output_fn)):
            os.makedirs(os.path.dirname(output_fn))
        if not os.path.exists(os.path.dirname(script_fn)):
            os.makedirs(os.path.dirname(script_fn))
        if not os.path.exists(os.path.dirname(log_filestem)):
            os.makedirs(os.path.dirname(log_filestem))
        fo = open(script_fn, 'w')
        print('''\
pysamstats --type coverage_binned \
--window-size=%d \
--chromosome %s \
--start %d \
--end %d \
--fasta /lustre/scratch109/malaria/pfalciparum/resources/3D7_V3.fasta \
%s > %s
''' % (
                exon_length,
                analysis_exons[exon][0],
                analysis_exons[exon][1],
                analysis_exons[exon][2],
                bam_fn,
                output_fn
            ),
            file = fo
        )
        fo.close()
        st = os.stat(script_fn)
        os.chmod(script_fn, st.st_mode | stat.S_IEXEC)
        bsub(
            '-G', 'malaria-dk',
            '-P', 'malaria-dk',
            '-q', 'normal',
            '-o', '%s.out' % (log_filestem),
            '-e', '%s.err' % (log_filestem),
            '-J', '%s_%s' % (exon, ox_code),
            '-R', "'select[mem>4000] rusage[mem=4000]'",
            '-M', '4000',
            script_fn)
    
```
